Query to join principals and episode where category = actor, ordered by ordering

```python
%%bigquery
select * from imdb_staging.title_principals as principals inner join imdb_staging.title_episode as episode on principals.tconst = episode.tconst where principals.category='actor' order by principals.ordering limit 10
```

Query to join basics and ratings where titleType = short, ordered by descending averageRating 

```python
%%bigquery
select basics.tconst, basics.titleType, basics.primaryTitle, ratings.averageRating, ratings.numVotes from imdb_staging.title_basics as basics join imdb_staging.title_ratings as ratings on basics.tconst=ratings.tconst where basics.titleType = 'short' order by averageRating desc limit 10
```

Query to join akas and principals where region = FT, ordered by asceding principals ordering

```python
%%bigquery
select * from imdb_staging.title_akas as akas join imdb_staging.title_principals as principals on akas.titleId=principals.tconst where akas.region='FR' order by principals.ordering limit 10
```

Query to join akas and episode where types = original, ordered by ascending title

```python
%%bigquery
select akas.region, AVG(ratings.averageRating) as Average_Rating, SUM(ratings.numVotes) as Total_Votes from imdb_staging.title_akas as akas join imdb_staging.title_ratings as ratings on akas.titleId=ratings.tconst group by akas.region order by Total_Votes desc limit 10
```

Query that finds the average of the average ratings and sums the number of votes for the rating. It joins akas and ratings, grouping by region and only displaying the votes that are greater than 1000. It is also ordered by the average rating.

```python
%%bigquery
select akas.region, AVG(ratings.averageRating) as Average_Rating, SUM(ratings.numVotes) as Total_Votes from imdb_staging.title_akas as akas join imdb_staging.title_ratings as ratings on akas.titleId=ratings.tconst group by akas.region having Total_Votes > 1000 order by Average_Rating  desc limit 10
```

Query to find the minimum first year for director, grouping by director and only displaying those that started later than 1930 ordering by director

```python
%%bigquery
select bollywood.Director, min(bollywood.Year) as First_Year from kaggle_staging.bollywood as bollywood group by bollywood.Director having First_Year > 1930 order by bollywood.Director limit 10
```

Query to count the height of actors grouping by actor height having the count greater than 10 and ordering by actor height

```python
%%bigquery
select actors.Height_in_cm_, count(actors.Height_in_cm_) as count from kaggle_staging.bollywood_actors as actors group by actors.Height_in_cm_ having count >= 10 order by Height_in_cm_ desc limit 10
```

Query to join bollywood and actresss where Genre=social, ordered by ascending Year

```python
%%bigquery
select * from kaggle_staging.bollywood as bollywood join kaggle_staging.bollywood_actress as actress on bollywood.Title=actress.Debut_aslead_role where bollywood.Genre='social' order by bollywood.Year limit 10
```

Query that counts the bollywood actress name joining bollywood and actress by grouping by actress name

```python
%%bigquery
select actress.Name, count(bollywood) as count from kaggle_staging.bollywood as bollywood join kaggle_staging.bollywood_actress as actress on bollywood.Title=actress.Debut_aslead_role group by actress.Name order by count desc limit 10
```

Query to that counts the bollywood titles for each director grouping by director and having the director not equal to unknown 

```python
%%bigquery
select bollywood.Director, count(bollywood.Title) as count from kaggle_staging.bollywood as bollywood group by bollywood.Director having bollywood.Director != 'Unknown' order by count desc limit 10
```
