Inner joining city with country schemas by country_id

Querying city id's of >500

Ordered by ascending city id's

```python
!psql -c "select * from pagila.city inner join pagila.country on city.country_id = country.country_id where city_id > 500 order by city_id limit 10"
```

Inner joining store with customer

Querying customer id's of >500

Ordered by descending customer id's

```python
!psql -c "select store.store_id, store.manager_staff_id, store.address_id, customer.customer_id, customer.first_name, customer.last_name from pagila.store inner join pagila.customer on store.store_id = customer.store_id where customer_id > 500 order by customer_id desc limit 10"
```

Inner joining inventory with rental
Querying by store id ascending equal to 2
Ordering by staff id 

```python
!psql -c "select * from pagila.inventory inner join pagila.rental on inventory.inventory_id = rental.inventory_id where store_id = 2 order by staff_id limit 10"
```

Left outer joining inventory with rental
Querying by store id ascending equal to 2
Ordering by staff id 

```python
!psql -c "select * from pagila.inventory left outer join pagila.rental on inventory.inventory_id = rental.inventory_id where store_id = 2 order by staff_id limit 10"
```

Right outer joining actor with film actor 
Querying by film id ascending equal to 3 
Ordering by film id 

```python
!psql -c "select * from pagila.actor right outer join pagila.film_actor on actor.actor_id = film_actor.actor_id where film_id = 3 order by film_id"
```

Full outer joining category with film category 
Querying by film id descending less than 10 
Ordering by film id 

```python
!psql -c "select * from pagila.category full outer join pagila.film_category on category.category_id = film_category.category_id where film_id < 10 order by film_id"
```

Self join address id with district id and city id 
<br>
Querying by address by id descending

```python

!psql -c "select a.address_id, b.address_id, a.district, b.city_id from pagila.address a, pagila.address b where a.address_id <> b.address_id and a.city_id = b.city_id order by a.address_id desc limit 10"
```
