```python
import pandas as pd
from google.cloud import firestore
db = firestore.Client()

## add this to your first notebook cell
import uuid

## inside your code to populate/create the collections
variable_name = str(uuid.uuid4())
```

```python
batch = db.batch()

df = pd.read_csv('/home/jupyter/snippets/spanner_shopify/apps.csv')
apps_rows = df.values.tolist()

i = 0

for apps_row in apps_rows:
    i += 1
    app_record = {}
    app_record['id'] = apps_row[0]
    app_record['url'] = apps_row[1]
    app_record['title'] = apps_row[2]
    app_record['developer'] = apps_row[3]
    app_record['developer_link'] = apps_row[4]
    app_record['icon'] = apps_row[5]
    app_record['rating'] = apps_row[6]
    app_record['reviews_count'] = apps_row[7]
    
    apps_ref = db.collection('apps').document(apps_row[0])
        
    
    batch.set(apps_ref, app_record)
    if i == 400:
        i = 0
        batch.commit()

batch.commit()
```

```python
batch = db.batch()

df = pd.read_csv('/home/jupyter/snippets/spanner_shopify/categories.csv')
category_rows = df.values.tolist()

df = pd.read_csv('/home/jupyter/snippets/spanner_shopify/apps_categories.csv')
app_category_rows = df.values.tolist()

df = pd.read_csv('/home/jupyter/snippets/spanner_shopify/apps.csv')
apps_rows = df.values.tolist()

i = 0

for category_row in category_rows:
    i += 1
    category_record = {}
    category_record['id'] = category_row[0]
    category_record['title'] = category_row[1]
    
    category_ref = db.collection('category').document(category_row[0])
    
    
    for app_category_row in app_category_rows:
        if app_category_row[1] == category_row[0]:
            j = 0
            for k in range(len(apps_rows)):
                if apps_rows[k][0] == app_category_row[0]:
                    j += 1
                    app_record = {}
                    app_record['url'] = apps_rows[k][1]
                    app_record['title'] = apps_rows[k][2]
                    app_record['developer'] = apps_rows[k][3]
                    app_record['developer_link'] = apps_rows[k][4]
                    app_record['icon'] = apps_rows[k][5]
                    app_record['rating'] = apps_rows[k][6]
                    app_record['reviews_count'] = apps_rows[k][7]
                    
                    apps_ref = category_ref.collection('apps').document(apps_rows[k][0])
                    
                    batch.set(apps_ref, app_record)
                    
                if j == 400 or k == len(apps_rows) - 1:
                    k = 0
                    batch.commit()
            batch.commit()
    
    batch.set(category_ref, category_record)
    
    if i == 400:
        i = 0
        batch.commit()

batch.commit()
```

```python
apps_ref = db.collection('apps')
results = apps_ref.stream()
count = 0

for result in results:
    count += 1

print(f"number of apps: {count}")
```

```python
categories_ref = db.collection('category')
results = categories_ref.stream()
count = 0
subcount = 0

for result in results:
    count += 1
    apps_ref = categories_ref.document(result.id).collection('apps').stream()
    for app_ref in apps_ref:
        subcount += 1

print(f"number of categories: {count}")
print(f"number of apps: {subcount}")
```
